```python
from pymodulon.core import IcaData
from pymodulon.plotting import *
from pymodulon.io import load_json_model
import pandas as pd

from os import path
import re
from tqdm.notebook import tqdm
from pymodulon.util import _parse_sample
```

```python
import plotly.graph_objects as go
import plotly.express as px
```

```python
plt.rcParams['svg.fonttype'] = 'none'
```

```python
ica_data = load_json_model('../../putidaPRECISE321.json')
```

```python
FleQ1=pd.DataFrame(pd.DataFrame(ica_data.view_imodulon("FleQ/Fur")['COG']).value_counts())
```

```python
FleQ2=pd.DataFrame(pd.DataFrame(ica_data.view_imodulon("FleQ/AmrZ")['COG']).value_counts())
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:

explode = (0.1, 0, 0, 0,0,0,0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(FleQ1[0], explode = explode, labels=FleQ1.index,
        autopct='%1.1f%%',
        shadow=False, startangle=90)

ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
fig1.savefig("./raw_files_new/FigS5b_FleQ_Fur_COG.svg")
```
