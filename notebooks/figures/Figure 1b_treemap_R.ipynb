```r
pdf(file = file.path('.','raw_files','Fig1b_treemap.pdf'),
    width = 7,
    height = 8)


treemap(data,
            index=c("Category","imodulon_name"),
            vSize="exp_var",
            type="categorical",
            vColor='Category',
            fontsize.labels=12,
            palette='Set3',
            )

dev.off()
```
