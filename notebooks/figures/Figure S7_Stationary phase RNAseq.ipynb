# Load metadata

## Reorder column and row names

```r
counts <- counts[,order(colnames(counts))] # Reorder sample names in count table
coldata <- coldata[order(rownames(coldata)),] # Reorder sample names in metadata table
                         
counts <- counts[order(rownames(counts)),] # Reorder genes in count table
```

```r
head(counts)
```

```r
coldata
head(coldata)
```

## Ensure that metadata and count data have same columns

```r
all(colnames(counts) == rownames(coldata))
```

## Select project
This is an important step. You only want to keep data from the project that you are testing DEGs. If you skip this step, it could take up to an hour to compute DEGs!

```r
project_name = 'Phase' # Change project name here
```

```r
proj_coldata <- coldata[coldata$project == project_name,]
proj_coldata
```

```r
proj_counts <- counts[,rownames(proj_coldata)]
proj_counts
```

# Create DESeq Data Set

```r
suppressPackageStartupMessages(library('DESeq2'))
```

```r
proj_coldata
```

```r
dds <- DESeqDataSetFromMatrix(countData = proj_counts,
                              colData = proj_coldata,
                              design = ~ condition)
```

```r
dds <- DESeq(dds)
dds
```

# Run differential expression
We will turn off the Cook's cutoff since we usually have 3 or fewer replicates (see [Documentation](https://www.bioconductor.org/packages/release/bioc/vignettes/DESeq2/inst/doc/DESeq2.html#Approach-to-count-outliers))

**A positive value in the DEG table means that sample 1 has higher expression of the gene than sample 2.**

```r
sample1 = 'WT_St' # Put the name of your condition here
sample2 = 'WT_Ex' # Put the name of your second condition here

res <- results(dds,contrast=c('condition',sample1,sample2),
               cooksCutoff=FALSE)
```

```r
deg_table <- res[complete.cases(res[,c('padj','log2FoldChange')]) & 
                 # ^ Make sure that LogFoldChange and p-adjusted are complete (not null)
                res$padj < .05 & # Adjusted p-value cutoff of 0.05
                abs(res$log2FoldChange) > 1,] # abs(log2foldChange) cutoff of 1
```

# Make a volcano plot

```r
library(EnhancedVolcano)
```

```r
pdf(file = './raw_files_new/FigS7_stationary_volcano.pdf')
EnhancedVolcano(res,lab=rownames(res),
                x='log2FoldChange',
                y='padj',
                # Change arguments below to edit cutoffs
                pCutoff = .05,
                FCcutoff=1.0,
                # Change arguments below to edit point size and fontsize
               # transcriptPointSize = 3.0,
                # transcriptLabSize = 4.0,
                axisLabSize=12.0,
                title='',
                subtitle='',
                captionLabSize=12.0,
                legendLabSize=12.0
               )
dev.off()
```
